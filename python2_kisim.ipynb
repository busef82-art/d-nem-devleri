```python
# KÜTÜPHANELER
import math  # #Kütüphane Kullanımı (math)
from datetime import datetime  # #Kütüphane Kullanımı (datetime)

# MÜŞTERİ LİSTESİ (5 kişi - dictionary + liste)
musteriler_listesi = [
    {"ad": "Ali", "aylik_ucret": 1000, "hizmet": "Internet"},
    {"ad": "Ayşe", "aylik_ucret": 1500, "hizmet": "TV"},
    {"ad": "Mehmet", "aylik_ucret": 1200, "hizmet": "Telefon"},
    {"ad": "Zeynep", "aylik_ucret": 2000, "hizmet": "Internet"},
    {"ad": "Berk", "aylik_ucret": 1800, "hizmet": "TV"}
]

# FONKSİYON TANIMLAMA
def tutar_hesapla(aylik_ucret):
    # #Fonksiyon Tanımlama
    kdvli_tutar = aylik_ucret * 1.20  # %20 KDV ekleme
    return kdvli_tutar


# TOPLAM FATURA HESAPLAMA
toplam = 0

# DÖNGÜ KULLANIMI
for musteri in musteriler_listesi:
    kdvli = tutar_hesapla(musteri["aylik_ucret"])
    toplam += kdvli
    print(musteri["ad"], "KDV'li tutar:", kdvli)

# math KÜTÜPHANESİ İLE YUVARLAMA
toplam_yuvarlanmis = math.ceil(toplam)

print("\nToplam Fatura:", toplam)
print("Yuvarlanmış Toplam Fatura:", toplam_yuvarlanmis)

# datetime KÜTÜPHANESİ
bugun = datetime.now()
print("Fatura Tarihi:", bugun)

# SET KULLANIMI (benzersiz hizmetler)
hizmet_listesi = []

for musteri in musteriler_listesi:
    hizmet_listesi.append(musteri["hizmet"])

benzersiz_hizmetler = set(hizmet_listesi)

print("\nBenzersiz Hizmetler:")
print(benzersiz_hizmetler)
```
