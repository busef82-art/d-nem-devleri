```python
import random

# 1. Değişkenler ve Tipler
ad_soyad = "Buse Fidan"
aylik_ucret = 650.0
sadakat_ayi = 30
aktif_mi = True

# 2. Liste ve Sözlük Kullanımı

# Şirketin sunduğu hizmetler (Liste)
hizmetler = ["İnternet", "Mobil Hat", "TV Paketi", "Bulut Depolama", "Güvenlik"]

# Müşteri bilgileri (Sözlük)
musteri = {
    "ad_soyad": ad_soyad,
    "aylik_ucret": aylik_ucret,
    "sadakat_ayi": sadakat_ayi,
    "aktif_mi": aktif_mi,
    "hizmetler": hizmetler
}

# 3. Koşullu İfadeler (If-Else)

if musteri["aylik_ucret"] > 500 or musteri["sadakat_ayi"] > 24:
    print("VIP Müşteri: İndirim Tanımla")
else:
    print("Standart Müşteri")

# 4. String Operasyonları

# İsmi büyük harfe çevirme
buyuk_isim = musteri["ad_soyad"].upper()

# Rastgele ID oluşturma
rastgele_sayi = random.randint(1000, 9999)
customer_id = "IST-2026-" + str(rastgele_sayi)

print("\n--- MÜŞTERİ BİLGİLERİ ---")
print("Ad Soyad:", buyuk_isim)
print("Customer ID:", customer_id)
print("Aylık Ücret:", musteri["aylik_ucret"])
print("Sadakat Ayı:", musteri["sadakat_ayi"])
print("Aktif mi:", musteri["aktif_mi"])
print("Hizmetler:", musteri["hizmetler"])
```
